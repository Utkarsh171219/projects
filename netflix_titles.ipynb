```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['title','type'])
df.head()
```

```python
df.isnull().sum()
```

```python
df['director'].fillna('unknown',inplace=True)
df['cast'].fillna('unkonown',inplace=True)
df['country'].fillna('unknown',inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'],format='mixed')
df.head()
```

```python
df['year_added']= df['date_added'].dt.year
df['month_added']= df['date_added'].dt.month
df[['date_added','year_added','month_added']].head()
```

```python
sns.countplot(data=df,x='type')
plt.title('Number of Movies vs TV Shows')
plt.xlabel('type')
plt.ylabel('count')
plt.show()
```

```python
df['year_added'].value_counts().sort_index().plot(kind='bar',figsize=(12,5))
plt.title('Contents added to the netflix by year')
plt.xlabel('year')
plt.ylabel('Number of titles added')
plt.show()
```

```python
top_countries=df['country'].value_counts().head(10)
top_countries.plot(kind='barh',figsize=(10,6),color='skyblue')
plt.title('top 10 countries with the most content')
plt.xlabel('number of titles')
plt.ylabel('country')
plt.gca().invert_yaxis()
plt.show()
```

```python
from collections import Counter

all_genres = df['listed_in'].dropna().str.split(',')
flat_genre_list=[genre for sublist in all_genres for genre in sublist]

genre_counts = Counter(flat_genre_list).most_common(10)

genre_df = pd.DataFrame(genre_counts,columns=['count','genre'])

sns.barplot(data=genre_df,x='count',y='genre',palette='pastel')
plt.title('top 10 most common genre on netflix')
plt.xlabel('count')
plt.ylabel('genre')

plt.show()
```
